# Exploratory Data Analysis of Steam's Video Games

This notebook explores the data collected from the Steam Web API and Steam Web Store. Included is an analysis of over 5000 video games hosted by Steam, and the game libraries of approximately 40,000 Steam users. This analysis helped in the production of LookingForSteamPlayers.com.

```python
from __future__ import division
import matplotlib.pyplot as plt
import matplotlib.axes as ax
import seaborn as sns
from datetime import datetime
sns.set_style("dark")

%pylab inline --no-import-all
```

Below, the video game data is loaded in from SQL where the columns include the game's name, developers, publishers, categories, genres, price, metacritic score, number of achievements, number of recommendations by Steam users, number of DLC (downloadable content), and release date. 

```python
# Pull data from MySQL
import sys
import MySQLdb as mdb
from dataCollecting.privateData import sqlpwd
from pandas import DataFrame, Series
import pandas as pd

sqlcon = mdb.connect('localhost','root',sqlpwd,'lfp')
gameData = pd.read_sql('SELECT * FROM gamedata', con = sqlcon,index_col = 'index')
gameData.original_price = gameData['original_price'].apply(lambda x: x/100) # Convert price to dollars
gameData.head()
```

## Overview of Games

```python
# Create new column identifying whether or not the game has multiplayer 

gameData['multi'] = 2  # Initiate all values to 2 
gameData.ix[(gameData['categories'].str.contains('Single-player')==True),'multi'] = 0

multiTypes = ['Multi-player','multiplayer','MMO','Co-op']
for mtype in multiTypes:
    gameData.ix[(gameData.categories.str.contains(mtype)==True),'multi'] = 1 # check for multi-player

invalidGameIDs = gameData[gameData.multi == 2].index
gameData = gameData[gameData.multi != 2]

# Generate some quick stats
nGames = len(gameData)
nFreeGames = len(gameData[gameData.is_free == True])

nMulti = len(gameData[gameData.multi == 1])
nSingle = len(gameData[gameData.multi == 0])

numberRatedM = gameData.metacritic.count()
numberRatedU = len(gameData[gameData.recommendations > 0])

noAchievements = (len(gameData[gameData.achievements== 0]))
nAchievements = (len(gameData[gameData.achievements > 0]))

nNoAgeRequirement = (len(gameData[gameData.age == '0']))
nAgeAdult = (len(gameData[gameData.age >= '17']))

print 'Quick Stats:'
print '\tThere are ' + str(nGames) + ' overall games on Steam, and ' + str(nFreeGames) + ' are free to play' 
print '\tThere are ' + str(nMulti) + ' multiplayer games currently available on Steam and ' + str(nSingle) + ' strictly single player games'
print '\tOnly '+str(int(100*numberRatedM/len(gameData)))+'% of games have been rated by metacritic, whereas '+str(int(100*numberRatedU/len(gameData)))+'% of games have been rated by users'
print '\tThere are ' + str(noAchievements) + ' games without Steam achievements, and ' + str(nAchievements) + ' games with Steam achievements' 
print '\tThere are ' + str(nNoAgeRequirement) + ' games without an age requirement, and ' + str(nAgeAdult) + ' games only for adults' 
```

## How many games are being released each month? 

```python
by = lambda x: lambda y: getattr(y,x)
startDate = '20060101'
endDate = '20150430'

plt.figure(figsize=(16,4),edgecolor='k',facecolor ='w');

plt.subplot(121)    
gameData.release_date.groupby(gameData.release_date.dt.year).count().plot(kind="bar",by='devRank')
plt.xlabel('Date',fontsize=14); plt.ylabel('# of Games Released',fontsize=14)

plt.subplot(122) 
recentReleases = gameData[(gameData.release_date > startDate) & (gameData.release_date < endDate)]
releases = Series(data= recentReleases.index,index = recentReleases.release_date) 
releaseCnts = Series(releases.groupby([by('year'),by('month')]).count())
releaseCnts.ix[2012:2015].plot(kind="bar")
plt.xlabel('Date',fontsize=14); plt.ylabel('# of Games Released',fontsize=14)

plt.suptitle('Increase in the Release of Games', fontsize = 16)
```

As seen above, 1817 video games were released in 2014; only 196 less than the total number of games released in all years prior. 

Beginning in 2012 and through the summer of 2013, the number of games released fluctuates around 30 per month and doesn't exceeds 50 releases in a month. However, near August of 2013 the release rate on average increases approximately 8 games per month and has yet to plateau. In 2014 and 2015 it appears that April is when the most games are released.   

### How are independent game developers contributing to the trend? 

```python
gameData['devRank'] = 'aaa'
gameData.ix[gameData['genres'].str.contains('Indie')==True,'devRank'] = 'indie' 

devs = ['aaa','indie']; devPloty = DataFrame(); devPlotm = DataFrame()

for dev in devs:
    temp = gameData[gameData.devRank==dev]
    temp1 = Series(temp.release_date.groupby(temp.release_date.dt.year).count())
    devPloty = pd.concat([devPloty, temp1], axis=1)
    
    recentReleases = gameData[(gameData.release_date > startDate) & (gameData.release_date < endDate) & (gameData.devRank ==dev)]
    releases = Series(data= recentReleases.index,index = recentReleases.release_date) 
    temp1 = Series(releases.groupby([by('year'),by('month')]).count())
    devPlotm = pd.concat([devPlotm, temp1], axis=1)
    
plt.figure(figsize=(16,4),edgecolor='k',facecolor ='w');

axs = plt.subplot(121)   
devPloty.columns = ['AAA','Indie']
devPloty.plot(kind='bar',stacked=True,ax=axs); plt.legend(loc=2,fontsize=14); 
plt.xlabel('Date',fontsize=14); plt.ylabel('# of Games Released',fontsize=14)

axs = plt.subplot(122)   
devPlotm.columns = ['AAA','Indie']
devPlotm.ix[2012:2015].plot(kind='bar',stacked=True,ax=axs); plt.legend(loc=2,fontsize=14); 
plt.xlabel('Month of Game Release',fontsize = 14); plt.ylabel('# of Games Released',fontsize = 14)

plt.suptitle('Increase in the Release of Games by Publishers', fontsize = 16)

devPlotm.plot(kind='line',figsize=(16,4)); plt.legend(loc=2,fontsize=14); 
plt.xlabel('Month of Game Release',fontsize = 14); plt.ylabel('# of Games Released',fontsize = 14)
```

Using Steam's "Indie" tag, and grouping all remaining games under the "AAA" umbrella - it is highly evident that indie game developers have had a significant contribution to the rise in release rate. Demonstrating that the gaming industry is experiencing a boom in independent game development, however there is also a general trend for increased game releases. One note to consider, is that the timing of this boom in independent game developers correlates with the recent boom in craft and microbreweries, which might demonstrate not only a shift in the gaming industry but perhaps a shift in consumer mindset.

# Incorporating User Data for Games

For this project, the game libraries of 50,000 steam users were pulled using the Steam Web API and users who did not have more that 10 hours of total game play were discounted, resulting in nearly 40,000 unique users being entered into the database owning 5742 games.  Note that this number of games is greater than the actual number of games on Steam due to some non-game items.  Therefore, the list is reduced based on the valid game ids. 

```python
# Pull user's game data from MySQL
sqlcon = mdb.connect('localhost','root',sqlpwd,'lfp')
usersGames = pd.read_sql('SELECT * FROM usersGameData', con = sqlcon,index_col = 'index')
usersGames.playtime = usersGames['playtime'].apply(lambda x: x/60)  # Convert minutes of game play to hours
usersGames.playtime_2weeks = usersGames['playtime_2weeks'].apply(lambda x: x/60)  # Convert minutes of game play to hours
sqlcon.close()

usersGames.head()
```

```python
nUsers = len(usersGames.user_id.unique())
nGamesCollected = len(usersGames.game_id.unique())
nUserGamePoints = len(usersGames)
print 'This database has ' + str(nUsers) + ' users, ' + str(nGamesCollected) + ' games, and ' + str(nUserGamePoints) + ' user/game combos'
```

```python
# Add user data to game data
groupedGames = usersGames.groupby('game_id')

gameData['n_owned'] = groupedGames.count()['user_id'] # Number of copies owned 
gameData['p_owned'] = 100*gameData['n_owned']/len(usersGames['user_id'].unique()) # Percentage of people who own a copy

gameData['playtime'] = groupedGames.sum()['playtime'] # Number of hours played all time
gameData['playtime_2weeks'] = groupedGames.sum()['playtime_2weeks'] # Number of hours played recently 

t = usersGames[usersGames['playtime'] == 0].groupby('game_id').count()
gameData['n_unplayed'] = t['user_id'] # Number of copies owned by never played
gameData['p_unplayed'] = 100*t['user_id']/gameData['n_owned'] # Percentage of copies which have never been played 

gameData.iloc[0:2]
```

```python
# Sanity Check - Check groupby accuracy using MySQL
sqlcon = mdb.connect('localhost','root',sqlpwd,'lfp')
testGames = pd.read_sql('SELECT * FROM usersGameData WHERE game_id = 10', con = sqlcon,index_col = 'index')
sqlcon.close()

print 'Number of Copies Owned = ' + str(len(testGames)) + ' from SQL and ' + str(int(gameData.ix[10]['n_owned'])) + ' from groupby'
print 'Number of Copies Unplayed = ' + str(len(testGames[testGames['playtime'] == 0])) + ' from SQL and ' + str(int(gameData.ix[10]['n_unplayed'])) + ' from groupby'
```

## Game Genres on Steam

All Steam games are categorized with at least one of 10 genre tags. Note that these genres are not mutually exclusive, such that a game can be an Indie-Action-Adventure game. Additionally, three of these genres do not describe the gameplay style, these are 'Early Access', 'Indie', and 'Free-to-Play.'  Note, free-to-play is not included in the subsequent analysis. 

```python
genreList = ['Action','Adventure','Casual','Early Access','Indie','Massively Multiplayer','Simulation','Sports','Strategy','Racing','RPG']

genrePlaytime = []; genreCnt = []; genreHrsPerDollar = []; genreAvgCost = []
i = 0
for genre in genreList:
    subset = gameData.ix[(gameData['genres'].str.contains(genre)==True)] 
    genreCnt.append(len(subset))
    genrePlaytime.append(subset.playtime.sum()/genreCnt[i]) 
    genreAvgCost.append(subset.original_price.sum()/genreCnt[i])
    genreHrsPerDollar.append(genrePlaytime[i]/genreAvgCost[i])
    i += 1
    
plt.figure(figsize=(15,9),edgecolor='k',facecolor ='w');
cmap = sns.color_palette("Blues_d", 12)

xlist = genreList[0:5] + ['MMO'] + genreList[6:]

plt.subplot(211)
plt.bar(range(len(genreList)), genreCnt, color=cmap,align='center')
plt.xticks(range(len(genreList)),xlist)
plt.title('Steam Game Genres',fontsize=14); plt.ylabel('Number of Games',fontsize=12); # plt.xlabel('Genres',fontsize=12)

plt.subplot(212)
plt.bar(range(len(genreList)), genrePlaytime, color=cmap,align='center')
plt.xticks(range(len(genreList)),xlist)
plt.title('Normalized Genre Playtime',fontsize=14); plt.ylabel('Playtime (Hours)',fontsize=12); 
```

Aside from Indie which is a genre based on the developer and not playstyle, Action is the most popular category by number of games in the Steam catalog. Classifications such a first person and third person shooters are wrapped into this Action category in Steam's database. Racing, sports, and massively multiplayer online games (MMOs) make up a relatively small segment of the catalog.  However, when normalizing user's total playtime in a genre by the number of games in that genre, it is clear that the MMOs averages longer gameplay per title. This is likely due to the open-ended nature of most MMO games. A similar argument can be made for the Action genre, where games such as Call of Duty can be played repeatedly in the online competitive mode. Additionally, many MMOs have a monthly subscription which incentivizes developers to constantly put out new content, keeping users engaged. Finally, this normalization reveals that although more than half of the games on Steam have an Indie genre tag, these games make up a small portion of the average playtime on Steam.  

```python
plt.figure(figsize=(15,9),edgecolor='k',facecolor ='w');
plt.subplot(211)
plt.bar(range(len(genreList)), genreAvgCost, color=cmap,align='center')
plt.xticks(range(len(genreList)),xlist)
plt.title('Average Game Cost by Genres',fontsize=14); plt.ylabel('Cost (USD)',fontsize=12); 

plt.subplot(212)
plt.bar(range(len(genreList)), genreHrsPerDollar, color=cmap,align='center')
plt.xticks(range(len(genreList)),xlist)
plt.title('Average "Bang-for-the-Buck"',fontsize=14); plt.ylabel('Average Playtime per USD',fontsize=12); plt.xlabel('Genres',fontsize=12)
```

For all genres, the average cost of a game on Steam is between \$8 and \$16, with Sports games averaging the highest price. Next, by normalizing playtime within a genre by the average cost, it is apparent that MMOs are the best "bang-for-the-buck". However, this does not account for monthly subscription fees which are common amongst MMOs. It should also be noted that microtransactions aren't taken into account, nor the cost of DLC or expansions, which can be significant.  

## Analysis of Metacritic Scores

Metacritic is a website that aggregates reviews for video games into a single score. Metacritic converts each review into a percentage and then takes a weighted average based on the critic's fame or stature. The video game breakdown for metacritic scores is, 

* 90-100: Universal acclaim 
* 75-89: Generally favorable
* 50-74: Mixed or average
* 20-49: Generally unfavorable
* 0-19: Overwhelming dislike

In this section, the Steam games which have metacritic scores are analyzed to see if any factors have an effect on a game's score. First, below the relationships between numerous game factors are summarized in a correlation plot. 

```python
rs = np.random.RandomState(33)
d = rs.normal(size=(100, 30))

f, ax = plt.subplots(figsize=(9, 9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.corrplot(gameData[['recommendations','achievements','dlc','metacritic','multi','original_price','p_owned','p_unplayed']], annot=False, sig_stars=False,
             diag_names=False, ax=ax, cmap=cmap)
f.tight_layout()
```

This plot shows that there is a high correlation between the number of achievements available for a game and the amount of DLC.  Additionally, metacritic score is not correlated with multi-player games, or Steam achievements, but is correlated slightly with price, Steam achievements, percentage of copies owned by users, and user recommendations. 

```python
bins = range(0,100,5)

plt.figure(figsize=(18,6),edgecolor='k',facecolor ='w');

axs = plt.subplot(121)
gameData[(gameData.multi==0)].metacritic.hist(alpha=0.5,bins=bins,ax=axs); 
gameData[(gameData.multi==1)].metacritic.hist(alpha=0.5,bins=bins,ax=axs)

plt.legend(['Single-player Games','Multi-player Games'],fontsize=14,loc=2)
plt.xlabel('Metacritic Score',fontsize=14); plt.ylabel('# of Games',fontsize=14)

axs = plt.subplot(122)
gameData[(gameData.original_price<=10)].metacritic.hist(alpha=0.4,bins=bins,ax=axs,color='crimson'); 
gameData[(gameData.original_price>10)].metacritic.hist(alpha=0.5,bins=bins,ax=axs,color='burlywood')

plt.legend(['Price <=$10','Price >$10'],fontsize=14,loc=2)
plt.xlabel('Metacritic Score',fontsize=14); plt.ylabel('# of Games',fontsize=14)
```

As seen in the correlation plot, multi-player games and single-player games have similar distributions. However, price does have a small effect on metacritic score, with less expensive games generally being rated lower.

#### Create Json file based on multi-player option to create D3 Bubbleplot

```python
from collections import defaultdict
import json 
import math

smallSet = gameData[['multi','name','p_owned']].copy()
smallSet1 = smallSet[smallSet['multi']==1].copy()
smallSet2 = smallSet[smallSet['multi']==0].copy()

jsonlist = {'name':'flare','children': []}
jsonlist['children'].append({'name': 'region1','children': []})
jsonlist['children'].append({'name': 'region2','children': []})

for i,(region,name,p_owned) in smallSet1.iterrows():
    if math.isnan(p_owned) == False:
        t = {'name': name, 'size': p_owned*100} 
        jsonlist['children'][0]['children'].append(t)
    
for i,(region,name,p_owned) in smallSet2.iterrows():
    if math.isnan(p_owned) == False:
        t = {'name': name, 'size': p_owned*100} 
        jsonlist['children'][1]['children'].append(t)
```

```python
with open('data/newBubble.json', 'w') as outfile:
    json.dump(jsonlist, outfile)
```
