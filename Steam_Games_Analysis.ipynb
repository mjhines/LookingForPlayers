## Exploratory Data Analysis of Steam's Video Games

This notebook explores the data collected from the Steam Web API and Steam Web Store. Included is an analysis of over 5000 video games hosted by Steam, and the game libraries of approximately 40,000 Steam users. This analysis helped in the production of LookingForSteamPlayers.com.

```python
from __future__ import division
import matplotlib.pyplot as plt
import matplotlib.axes as ax
import seaborn as sns
from datetime import datetime
sns.set_style("dark")
%pylab inline --no-import-all
```

Load in video game data from SQL where the columns include the game's name, develoeprs, publishers, categories, genres, price, metacritic score, number of achievements, number of recommendations by Steam users, number of DLC (downloadable content), and release date. 

```python
# Pull data from MySQL
import sys
import MySQLdb as mdb
from dataCollecting.privateData import sqlpwd
from pandas import DataFrame, Series
import pandas as pd

sqlcon = mdb.connect('localhost','root',sqlpwd,'lfp')
gameData = pd.read_sql('SELECT * FROM gamedata', con = sqlcon,index_col = 'index')
gameData.head()
```

### How many games have a multi-player feature? How many are free? How many are scored on metacritic? 

```python
# Create new column identifying whether or not the game has multiplayer 

gameData['multi'] = 2  # Initiate all values to 2 
gameData.ix[(gameData['categories'].str.contains('Single-player')==True),'multi'] = 0

multiTypes = ['Multi-player','multiplayer','MMO','Co-op']
for mtype in multiTypes:
    gameData.ix[(gameData['categories'].str.contains(mtype)==True),'multi'] = 1 # check for multi-player

invalidGameIDs = gameData[gameData['multi'] == 2].index
gameData = gameData[gameData['multi'] != 2]
```

```python
nGames = len(gameData)
nFreeGames = len(gameData[gameData['is_free'] == True])
print 'There are ' + str(nGames) + ' overall games on Steam, and ' + str(nFreeGames) + ' are free to play' 

nMulti = len(gameData[gameData['multi'] == 1])
nSingle = len(gameData[gameData['multi'] == 0])
print 'There are ' + str(nMulti) + ' multiplayer games currently available on Steam and ' + str(nSingle) + ' strictly single player games'

numberRatedM = gameData.metacritic.count()
numberRatedU = len(gameData[gameData['recommendations'] > 0])
print 'Only '+str(int(100*numberRatedM/len(gameData)))+'% of games have been rated by metacritic, whereas '+str(int(100*numberRatedU/len(gameData)))+'% of games have been rated by users'

noAchievements = (len(gameData[gameData['achievements'] == 0]))
nAchievements = (len(gameData[gameData['achievements'] > 0]))
print 'There are ' + str(noAchievements) + ' games without Steam achievements, and ' + str(nAchievements) + ' games with Steam achievements' 

nNoAgeRequirement = (len(gameData[gameData['age'] == '0']))
nAgeAdult = (len(gameData[gameData['age'] >= '17']))
print 'There are ' + str(nNoAgeRequirement) + ' games without an age requirement, and ' + str(nAgeAdult) + ' games only for adults' 
```

### How many games are being released each month? 

```python
by = lambda x: lambda y: getattr(y,x)
startDate = '20130101'
endDate = '20150430'

plt.figure(figsize=(16,4),edgecolor='k',facecolor ='w');

plt.subplot(121)    
gameData['release_date'].groupby(gameData['release_date'].dt.year).count().plot(kind="bar",by='devRank')
plt.xlabel('Date',fontsize=14); plt.ylabel('# of Games Released',fontsize=14)

plt.subplot(122) 
recentReleases = gameData[(gameData['release_date'] > startDate) & (gameData['release_date'] < endDate)]
releases = Series(data= recentReleases.index,index = recentReleases['release_date']) 
releases.groupby([by('year'),by('month')]).count().plot(kind="bar")
plt.xlabel('Date',fontsize=14); plt.ylabel('# of Games Released',fontsize=14)

plt.suptitle('Increase in the Release of Games', fontsize = 16)
```

As seen above, over 1817 video games were released in 2014; only 196 less than the number of games released in all years prior. 

Focusing only on games released after 2013 and binning game releases based on the month of release shows that the number of games released began to increase substantially at the end of 2013 - with more than double the number of games released in October and November of 2013 compared to the first 8 months of the year.  In 2014 and 2015 it appears that April is when the most games are released.   

###  How much are indie game developers participating to the trend? 

Dividing the data into the largest publishers, mid-size publishers, and indie publishers the data shows that some of the increase in game releases is due to 

```python
aaaDevs = ['Nintendo','Ubisoft','Electronic Arts','Sony','SQUARE ENIX','Microsoft','Bandai','Activision']
midDevs = ['Valve','Telltale','Paradox','Capcom','Take-Two','Sega','Zen','Devolver',
           'Konami','Slitherine','NIS','Warner Bros.','Koei','Atlus','505','Aksys','Deep','Focus','Disney','Lucasfilm'];

gameData['devRank'] = 'indie'
for dev in aaaDevs:
    gameData.ix[gameData['publishers'].str.contains(dev)==True,'devRank'] = 'aaa' 
for dev in midDevs:
    gameData.ix[gameData['publishers'].str.contains(dev)==True,'devRank'] = 'mid'     
```

```python
devs = ['aaa','mid','indie']; 
devPloty = DataFrame()
devPlotm = DataFrame()

for dev in devs:
    temp = gameData[gameData['devRank']==dev]
    temp1 = Series(temp['release_date'].groupby(temp['release_date'].dt.year).count())
    devPloty = pd.concat([devPloty, temp1], axis=1)
    
    recentReleases = gameData[(gameData['release_date'] > startDate) & (gameData['release_date'] < endDate) & (gameData['devRank']==dev)]
    releases = Series(data= recentReleases.index,index = recentReleases['release_date']) 
    temp1 = Series(releases.groupby([by('year'),by('month')]).count())
    devPlotm = pd.concat([devPlotm, temp1], axis=1)
    
plt.figure(figsize=(16,4),edgecolor='k',facecolor ='w');

axs = plt.subplot(121)   
devPloty.columns = ['AAA','Mid-Level','Indie']
devPloty.plot(kind='bar',stacked=True,ax=axs);
plt.xlabel('Date',fontsize=14); plt.ylabel('# of Games Released',fontsize=14)

axs = plt.subplot(122)   
devPlotm.columns = ['AAA','Mid-Level','Indie']
devPlotm.plot(kind='bar',stacked=True,ax=axs); plt.legend(loc=2); 
plt.xlabel('Month of Game Release',fontsize = 14); plt.ylabel('# of Games Released',fontsize = 14)

plt.suptitle('Increase in the Release of Games by Publishers', fontsize = 16)
```

### Analysis on Metacritic Scores

Metacritic is a website that aggregrates reviews for video games into a single score. Metacritic converts each review into a percentage and then takes a weighted average based on the criticc's fame or stature. The video game breakdown for metacritic scores is, 

90-100: Universal acclaim 

75-89: Generally favorable

50-74: Mixed or average

20-49: Generally unfavorable

0-19: Overwhelming dislike

Overall, multi-player games and single-player games have similar distributions when it comes to metacritic score.

```python
bins = range(0,100,5)

plt.figure(figsize=(10,6),edgecolor='k',facecolor ='w');

gameData[(gameData['multi']==0)]['metacritic'].hist(alpha=0.5,bins=bins); 
gameData[(gameData['multi']==1)]['metacritic'].hist(alpha=0.5,bins=bins)

plt.legend(['Single-player Games','Multi-player Games'],fontsize=14,loc=2)
plt.xlabel('Metacritic Score',fontsize=14); plt.ylabel('Frequency',fontsize=14)
```

# Incorporating User Data for Games

```python
# Pull user's game data from MySQL
sqlcon = mdb.connect('localhost','root',sqlpwd,'lfp')
usersGames = pd.read_sql('SELECT * FROM usersGameData', con = sqlcon,index_col = 'index')
sqlcon.close()

usersGames.head()
```

```python
nUsers = len(usersGames.user_id.unique())
nGamesCollected = len(usersGames.game_id.unique())
nUserGamePoints = len(usersGames)
print 'This database has ' + str(nUsers) + ' users, ' + str(nGamesCollected) + ' games, and ' + str(nUserGamePoints) + ' user/game combos'
```

```python
# Add user data to game data
groupedGames = usersGames.groupby('game_id')

gameData['n_owned'] = groupedGames.count()['user_id'] # Number of copies owned 
gameData['p_owned'] = 100*gameData['n_owned']/len(usersGames['user_id'].unique()) # Percentage of people who own a copy

gameData['playtime'] = groupedGames.sum()['playtime'] # Number of minutes played all time
gameData['playtime_2weeks'] = groupedGames.sum()['playtime_2weeks'] # Number of minutes played recently 

t = usersGames[usersGames['playtime'] == 0].groupby('game_id').count()
gameData['n_unplayed'] = t['user_id'] # Number of copies owned by never played
gameData['p_unplayed'] = 100*t['user_id']/gameData['n_owned'] # Percentage of copies which have never been played 

gameData.iloc[0:2]
```

#### Create Json file based on multi-player option to create D3 Bubbleplot

```python
from collections import defaultdict
import json 
import math

smallSet = gameData[['multi','name','p_owned']].copy()
smallSet1 = smallSet[smallSet['multi']==1].copy()
smallSet2 = smallSet[smallSet['multi']==0].copy()

jsonlist = {'name':'flare','children': []}
jsonlist['children'].append({'name': 'region1','children': []})
jsonlist['children'].append({'name': 'region2','children': []})

for i,(region,name,p_owned) in smallSet1.iterrows():
    if math.isnan(p_owned) == False:
        t = {'name': name, 'size': p_owned*100} 
        jsonlist['children'][0]['children'].append(t)
    
for i,(region,name,p_owned) in smallSet2.iterrows():
    if math.isnan(p_owned) == False:
        t = {'name': name, 'size': p_owned*100} 
        jsonlist['children'][1]['children'].append(t)
```
