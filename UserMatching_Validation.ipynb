# Validation for User Matching

In this program, the LookingForSteamPlayers user matching system is validated by comparing the average similarity scores between users and random users, users and their friends, and users and their matches. 

```python
from __future__ import division
import pickle
from pandas import DataFrame, Series
import pandas as pd
from dataCollecting.privateData import pdir
import random
import userMatch as um
import dataCollecting.collect as co
import numpy as np
import seaborn as sns
from scipy import stats

%pylab inline --no-import-all
```

This analysis is validated by selecting a random subset of users from a set of 40,000 Steam users. 

```python
# Import user-game data 
with open(pdir + 'data/usersGamesDict', 'rb') as f:
    userGames = pickle.load(f)
```

```python
# Take random set of users 
randomUserTestSet = random.sample(userGames.keys(),500)
randomUserCompareSet = random.sample(userGames.keys(),5000)
```

```python
# Find average similarity score for random user selection
matchedScores = []
for user in randomUserTestSet:
    userScores = um.topMatches(userGames,user,randomUserCompareSet, n=len(userGames))
    matchedScores.append([s[0] for s in userScores])
```

```python
# Find average similarity score for random user against their friends (given they have at least 5 friends)
friendScores = []
fewFriendScores = []
for user in randomUserTestSet:
    normalizeFactor = um.sim_cosine(userGames,user,user)
    if normalizeFactor > 0:
        try:
            # If the user's friends list is available, calculate their cosine similarity with each friend
            pFriends = co.getFriendList(user) 
            pFriends2 = co.listToDict(pFriends,'steamid') 
            friendCompare = [key.encode('ascii') for key in pFriends2.keys()]
            t = [um.sim_cosine(userGames,user,other)/normalizeFactor for other in friendCompare if other in userGames.keys()]
            if len(t) > 5:
                friendScores.append(sum(t)/float(len(t)))
            else:
                fewFriendScores.append(sum(t)/float(len(t)))
        except:
            pass
```

### Graphical comparison of distributions for average similarity score when paired with a random user, a friend, or top 10 from algorithm

```python
randomScores = [sum(m)/len(m) for m in matchedScores]
myScores = [sum(m[0:9])/10 for m in matchedScores]

rs = [item*10 for item in randomScores]
fs = [item*10 for item in friendScores]
ms = [item*10 for item in myScores]

sns.set(rc={"figure.figsize": (10, 6)})

bins = np.linspace(0, 10, 10)

cc = sns.color_palette("Set1", 3)
sns.distplot(Series(rs), bins=bins, color=cc[0],label='Avg. Random Pairing')
sns.distplot(Series(fs), bins=bins, color=cc[1],label='Avg. Friend Pairing')
sns.distplot(Series(ms), bins=bins, color=cc[2],label='Avg. Algorithm Pairing')

plt.legend(fontsize = 14,loc=9)
plt.xlim([-1,11])

sns.axlabel('Similarity Score', 'Normalized Number of Users',fontsize = 14)
```

## Random user distribution comparisons

```python
# Find average similarity score for random user selection
randomMatchScores = []
randomFriendScores = []
randomUserScores = []

for i in np.arange(0,100):
    # Select a random user sample and a random user
    randomUserCompareSet = random.sample(userGames.keys(),1000)
    user = random.sample(userGames.keys(),1)[0]
    
    # If they have friends analyze their similarities 
    try:
        normalizeFactor = um.sim_cosine(userGames,user,user)
        
        pFriends = co.getFriendList(user)
        pFriends2 = co.listToDict(pFriends,'steamid')
        friendCompare = [key.encode('ascii') for key in pFriends2.keys()]
        t = [um.sim_cosine(userGames,user,other)/normalizeFactor for other in friendCompare if other in userGames.keys()]
        randomFriendScores.append(random.sample(t,1)[0])
        
        temp = um.topMatches(userGames,user,randomUserCompareSet, n=len(userGames))
        randomUserScores.append(random.sample(temp,1)[0][0])
        randomMatchScores.append(temp[0][0])      
    except:
        pass   
```

```python
rs = [item*10 for item in randomUserScores]
fs = [item*10 for item in randomFriendScores]
ms = [item*10 for item in randomMatchScores]
```

```python
sns.set(rc={"figure.figsize": (10, 6)})

bins = np.linspace(0, 10, 10)

cc = sns.color_palette("Set1", 3)
sns.distplot(Series(rs), bins=bins, color=cc[0],label='Random Pairing')
sns.distplot(Series(fs), bins=bins, color=cc[1],label='Friend Pairing')
sns.distplot(Series(ms), bins=bins, color=cc[2],label='Algorithm Pairing')

sns.axlabel('Similarity Score', 'Normalized Number of Users')
```

### Visualization of 100 Users Similarity Scores

```python
def allMatches(prefs,user,compareKeys,n=5,similarity=um.sim_cosine):
    
    normalizeFactor = similarity(prefs,user,user)
    scores=[(similarity(prefs,user,other)/normalizeFactor,other) for other in compareKeys]
    
    # Sort the list so the highest scores appear at the top
    return scores    
```

```python
import seaborn as sns
import userMatch as um
```

```python
# Take random set of users 
randomUserTestSet = random.sample(userGames.keys(),100)

# Find average similarity score for random user selection
matchedScores = []
for user in randomUserTestSet:
    userScores = allMatches(userGames,user,randomUserTestSet)
    matchedScores.append([s[0] for s in userScores])

sortedScores = [sorted(m,reverse=True) for m in matchedScores]
avgSims = [np.mean(m) for m in matchedScores]
sortedSortedMatches = [x for (y,x) in sorted(zip(avgSims,sortedScores),reverse=True)]
```
