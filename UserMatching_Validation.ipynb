In the above heatmap, the full similarity matrix can be observed. Here, each row and column corresponds to a user. In this version, it is easier to identify users who matched well with a wide variety of people (the darker colors extend far out) as well as users who didn't match well with anyone (the entire row is of a lighter color). 

Finally, let's take the average similarity coeffient for each row and then rank the rows with the first row having the highest average similarity and the bottom row having the lowest average similarity. 

```python
# Sort rows so that first row has user with highest average similarity score
avgSims = [np.mean(m) for m in matchedScores] # find average similarity scores for each row
sortedSortedMatches = [x for (y,x) in sorted(zip(avgSims,sortedScores),reverse=True)] 

plt.figure(figsize=(14,11))

sns.heatmap(sortedSortedMatches, vmax=1.0, linewidths=0, square=True)
plt.tick_params(bottom='on',left='on',labelbottom='off',labelleft='off')
```

Again this plot shows a few different distribution types. The people towards the bottom don't tend to match well with many people. However, the people near the top tend towards two separate groups. One group has strong similarity with a few users whereas the another group has weaker similarities but with more users.

## Validation: Comparing Similarity Scores between Random Users, Friends, and Top Matches

Two steps were required in order to validate the proposed user matching system. First, it must be shown that the similarity score is valuable for predicting Steam user's enjoyment playing with one another. This is done by comparing the average similarity scores between random users with the average similarity scores between Steam friends. Next, it must also be shown that by ranking the similarity scores that the proposed algorithm performs better than random.    

```python
# Find average similarity score for random user selection
matchedScores = []; nGamesCommon = []; playtimeCommon = []
for user in randomUserTestSet:
    userScores = um.topMatches(userGames,user,randomUserCompareSet, n=len(userGames))
    matchedScores.append([s[0] for s in userScores])
    nGamesCommon.append([s[1] for s in userScores])
    playtimeCommon.append([s[2] for s in userScores])

# Find average similarity score for random user against their friends (given they have at least 5 friends)
friendScores = []; fewFriendScores = []
for user in randomUserTestSet:
    normalizeFactor = um.sim_cosine(userGames,user,user)[0]
    if normalizeFactor > 0:
        try:
            # If the user's friends list is available, calculate their cosine similarity with each friend
            pFriends = co.getFriendList(user) 
            pFriends2 = co.listToDict(pFriends,'steamid') 
            friendCompare = [key.encode('ascii') for key in pFriends2]
            t = [um.sim_cosine(userGames,user,other,normalizeFactor)[0] for other in friendCompare if other in userGames]
            if len(t) > 5:
                friendScores.append(sum(t)/float(len(t)))
            else:
                fewFriendScores.append(sum(t)/float(len(t)))
        except:
            pass
        
randomScores = [sum(m)/len(m) for m in matchedScores]
myScores = [sum(m[0:9])/10 for m in matchedScores]

# Scale similarity scores to be on a 1-10 basis 
rs = [item*10 for item in randomScores]
fs = [item*10 for item in friendScores]
ms = [item*10 for item in myScores]
```

### Graphical comparison of distributions for average similarity score when paired with a random user, a friend, or top 10 from algorithm

Here it is clear that the similarity score for friends is, on average, higher than that for random selected at random. Additionaly, the similarity score for a matched pairing averages twice as high of that of an average friend pairing!  

### Using SVD instead of Nearest Neighbor

Often times for recommendation systems matrix factorization is used in order to reduce the dimensionality of the problem to $k$ number of features. For user matching, the user-feature matrix can be determined and then users can be matched using cosine similarity of their interest in game features. This method is explored below.  

```python
from recsys.datamodel.data import Data
import recsys.algorithm
from recsys.algorithm.factorize import SVD
data = Data()
svd = SVD()
```

```python
with open(pdir + 'data/usersGamesDict', 'rb') as f:
    userGames = pickle.load(f)

# flatUserData = DataFrame(list(co.flatten_dict(userGames)))
# flatUserData.columns = ['user_id','game_id','playtime','playtime_2weeks']
# flatUserData[['user_id']] = flatUserData[['user_id']].astype(int)
# dataset = flatUserData[['playtime','user_id','game_id']].values
# dataset = [tuple(l) for l in dataset]
```

```python
# svd.set_data(dataset)
svd.load_model('data/svdData.zip')

# svd.compute(k=50, min_values=1, pre_normalize=None, mean_center=False, post_normalize=True,savefile='data/svdData')
```

After loading in the SVD model, the same method of computing random user pairings, friend pairings, and algorithm pairings is explored. Note that the random subset from the previous analysis is being used. 

```python
# Find average similarity score for random user selection
matchedScoresSVD = []
nGamesCommonSVD = []
userMatchesSVD = []
for user in randomUserTestSet:
    try:
        userScores = svd.similar(int(user),n=len(userGames))
        matchedScoresSVD.append([s[1] for s in userScores[1:] if str(s[0]) in randomUserCompareSet])
        userMatchesSVD.append([s[0] for s in userScores[1:] if str(s[0]) in randomUserCompareSet])
    except:
        pass

    
# Find average similarity score for random user against their friends (given they have at least 5 friends)
friendScoresSVD = []
fewFriendScoresSVD = []
for user in randomUserTestSet:
    try:
        # If the user's friends list is available, calculate their cosine similarity with each friend
        pFriends = co.getFriendList(str(user)) 
        pFriends2 = co.listToDict(pFriends,'steamid') 
        friendCompare = [int(key) for key in pFriends2]
        
        t = [svd.similarity(int(user),other) for other in friendCompare if str(other) in userGames]
        
        if len(t) > 1:
            friendScoresSVD.append(sum(t)/float(len(t)))
        else:
            fewFriendScoresSVD.append(sum(t)/float(len(t)))
    except:
        pass  
    
randomScoresSVD = [sum(m)/len(m) for m in matchedScoresSVD]
matchScoresSVD = [sum(m[0:9])/10 for m in matchedScoresSVD]

rs_svd = [item*10 for item in randomScoresSVD]
fs_svd = [item*10 for item in friendScoresSVD]
ms_svd = [item*10 for item in matchScoresSVD]
```

```python
sns.set(rc={"figure.figsize": (10, 6)})

bins = np.linspace(0, 10, 10)

cc = sns.color_palette("Set1", 3)
sns.distplot(Series(rs_svd), bins=bins, color=cc[0],label='Avg. Random Pairing')
sns.distplot(Series(fs_svd), bins=bins, color=cc[1],label='Avg. Friend Pairing')
sns.distplot(Series(ms_svd), bins=bins, color=cc[2],label='Avg. Algorithm Pairing')

plt.legend(fontsize = 14,loc=9)
plt.xlim([-1,11])

sns.axlabel('Similarity Score', 'Normalized Number of Users',fontsize = 14)
```

As with the nearest neighbor method, the similarity scores when matching users with random users, friends, and top algorithm candidates shows three distinct peaks, with the algorithm providing the greatest similarity. 

### Which method should be used? 

Given that both the nearest neighbor and SVD algorithms performed well on validation, the next question is which method should be used?  At this point there are two viable methods, and each method has its own advantages and disadvantages.. In both cases cosine similarity is required, and in the SVD user-feature method, the length of each user vector is smaller, however SVD also comes at the cost of an expensive computation that must be updated weekly.

This section is interested in detemrining if there is any critical reason not to simply chose one of these algorithms. 

First let's compare all of the similarity score distriubtions. 

```python
randomScoresSVD = [sum(m)/len(m) for m in matchedScoresSVD]
matchScoresSVD = [sum(m[0:9])/10 for m in matchedScoresSVD]

rs_svd = [item*10 for item in randomScoresSVD]
fs_svd = [item*10 for item in friendScoresSVD]
ms_svd = [item*10 for item in matchScoresSVD]

sns.set(rc={"figure.figsize": (10, 6)})

bins = np.linspace(0, 10, 10)

cc = sns.color_palette("hls", 8)
sns.distplot(Series(rs), bins=bins, color=cc[0],label='Avg. Random Pairing')
sns.distplot(Series(fs), bins=bins, color=cc[1],label='Avg. Friend Pairing')
sns.distplot(Series(ms), bins=bins, color=cc[2],label='Avg. Algorithm Pairing')

sns.distplot(Series(rs_svd), bins=bins, color=cc[3],label='Avg. Random Pairing SVD')
sns.distplot(Series(fs_svd), bins=bins, color=cc[4],label='Avg. Friend Pairing SVD')
sns.distplot(Series(ms_svd), bins=bins, color=cc[5],label='Avg. Algorithm Pairing SVD')

plt.legend(fontsize = 14,loc=9)
plt.xlim([-1,11])

sns.axlabel('Similarity Score', 'Normalized Number of Users',fontsize = 14)
```

Both the nearest neighbor and SVD methods seem to have very similar distributions. Let's instead increase the number of bins and look only at friend pairings and algorithm pairings.

It seems that the SVD method resulted in slightly less variation for the top algorithm pairings, with the lowest similarity score being about 6, whereas the nearest neighbor algorithm resulted in an average simlarity score as low as 2. 

There is also another issue to consider, users want to play games with people who own the same games as them and put similar amounts of time into those games. If in the SVD method users are being matched on features, it is possible that top matches won't share the same games -- but rather share the same taste in games. This can be evaluated by simply counting the number of games and number of minutes of gameplay that top matches have in common. 

```python
# Define a fuction to count common games and playtime 

def cntCommon(prefs,p1,p2):
    si={}
    n = 0
    commonPlaytime = 0
    for item in prefs[p1]:
        if (isinstance(item,int)) and (item in prefs[p2]):
            n+=1   
            si[item]=1
            if prefs[p1][item][0] > prefs[p2][item][0]:
                commonPlaytime = commonPlaytime + prefs[p1][item][0]
            else:
                commonPlaytime = commonPlaytime + (prefs[p2][item][0]-prefs[p1][item][0])

    return (n,commonPlaytime)
```

```python
ncommonsvd = []; nplaytimesvd = []
for user in randomUserTestSet:
    temp = [cntCommon(userGames,str(user),str(s)) for s in userMatchesSVD[0]]
    ncommonsvd.append([t[0] for t in temp])
    nplaytimesvd.append([t[1] for t in temp])
```

```python
k = 99

nCommon = [sum(m[0:k])/(k+1) for m in nGamesCommon]
nPlaytimeCommon = [sum(m[0:k])/(k+1) for m in playtimeCommon]

nCommonSVD = [sum(m[0:k])/(k+1) for m in ncommonsvd]
nPlaytimeCommonSVD = [sum(m[0:k])/(k+1) for m in nplaytimesvd]

sns.set(rc={"figure.figsize": (18, 6)})
cc = sns.color_palette("Set1", 6)

plt.subplot(121)
bins = np.linspace(0, 60, 20); cc = sns.color_palette("Set1", 3)
sns.distplot(Series(nCommon), bins=bins, color=cc[0],label='Game Matching')
sns.distplot(Series(nCommonSVD), bins=bins, color=cc[1],label='Feature Matching')
plt.legend(fontsize = 14,loc=1)
sns.axlabel('# of Games in Common', 'Normalized # of Users',fontsize = 14)

plt.subplot(122)
bins = np.linspace(0, 600000, 20); cc = sns.color_palette("Set1", 3)
sns.distplot(Series(nPlaytimeCommon), bins=bins, color=cc[0],label='Game Matching')
sns.distplot(Series(nPlaytimeCommonSVD), bins=bins, color=cc[1],label='Feature Matching')
plt.legend(fontsize = 14,loc=1)
sns.axlabel('Avg. Playtime in Common (Minutes)', 'Normalized # of Users',fontsize = 14)

# # plt.xlim([-1,11])
```

Finally, we have shown why we should use the original nearest neighbor algorithm. By matching based on playtimes in games, the number of games that the average matched pairing has in common has a greater chance of being more than 10. Additionally, when matched based on features instead of games, the highest chance for number of minutes of playtime together is the lowest bin.  Since users are looking for people to play the games they own with, matching based on features would be unreliable in providing this service.   

## Random user distribution comparisons

Finally, the same validation plot is determined using another random analysis where instead of comparing each user to a random subset of users, each random user is simply compared with one other random user. This method resulted in lower similarity scores for random-random user pairings, stronger matched pairings and the resulting user-friend pairing distribution (where the friend is selected at random) shows more variance than in the previous analysis. 

```python
# Find average similarity score for random user selection
randomMatchScores = []
randomFriendScores = []
randomUserScores = []

for i in np.arange(0,100):
    # Select a random user sample and a random user
    randomUserCompareSet = random.sample(userGames.keys(),1000)
    user = random.sample(userGames.keys(),1)[0]
    
    # If they have friends analyze their similarities 
    try:
        normalizeFactor = um.sim_cosine(userGames,user,user)
        
        pFriends = co.getFriendList(user)
        pFriends2 = co.listToDict(pFriends,'steamid')
        friendCompare = [key.encode('ascii') for key in pFriends2.keys()]
        t = [um.sim_cosine(userGames,user,other)/normalizeFactor for other in friendCompare if other in userGames.keys()]
        randomFriendScores.append(random.sample(t,1)[0])
        
        temp = um.topMatches(userGames,user,randomUserCompareSet, n=len(userGames))
        randomUserScores.append(random.sample(temp,1)[0][0])
        randomMatchScores.append(temp[0][0])      
    except:
        pass   

rs = [item*10 for item in randomUserScores]
fs = [item*10 for item in randomFriendScores]
ms = [item*10 for item in randomMatchScores]

sns.set(rc={"figure.figsize": (10, 6)})

bins = np.linspace(0, 10, 10)

cc = sns.color_palette("Set1", 3)
sns.distplot(Series(rs), bins=bins, color=cc[0],label='Random Pairing')
sns.distplot(Series(fs), bins=bins, color=cc[1],label='Friend Pairing')
sns.distplot(Series(ms), bins=bins, color=cc[2],label='Algorithm Pairing')

sns.axlabel('Similarity Score', 'Normalized Number of Users')
```
