## Game Recommendations using Scores Generated by Relative Gameplay

In this section each user's playtime is "scored" using quantiles. Using these scores, a utility matrix relating the users and their games is created, and divided into a training and test set (80/20). Next a SVD with 50 features is calculated using the python-recsys package https://github.com/ocelma/python-recsys and the normalized RMSE is determined. 

```python
sqlcon = mdb.connect('localhost','root',sqlpwd,'lfp')
allGames = pd.read_sql('SELECT game_id,user_id,playtime FROM usersGameData',con = sqlcon,index_col = 'game_id')
gameNames = pd.read_sql('SELECT * FROM gamedata', con = sqlcon,index_col = 'index')
sqlcon.close()

allGames = allGames.join(gameNames[['name']])
allGames.dropna(inplace=True)
allGames.reset_index(inplace=True)
allGames.rename(columns={'index':'game_id'}, inplace=True)
allGames = allGames[(allGames.playtime != 0)]
allGames.head()
```

```python
# Score every game using quantiles, starting with 10. 
# If there are not enough data points to support deciles, then reduce the amount of quantiles. 
def f(column):
    qmax = 10
    try:
        q = pd.qcut(column, qmax, labels=range(1,qmax+1))
    except ValueError: 
        try:
            q = pd.qcut(column, 5, labels=[2,4,6,8,10])
        except:
            try:
                q = pd.qcut(column, 3, labels=[2,5,8])
            except:
                q = 5                
    return q
    
allGames['score'] = allGames.groupby(['game_id'])['playtime'].transform(f)
allGames.head()
```

```python
# Format dataset into tuples and assign to recsys data format 
from recsys.datamodel.data import Data
data = Data()
dataset = allGames[['score','user_id','game_id']].values
dataset = [tuple(l) for l in dataset]
data.set(dataset)
```

```python
# Divide data into training and test set 
train, test = data.split_train_test(percent=80)
```

```python
# Import svd algorithm and calculate using training set and 50 features 
import recsys.algorithm
from recsys.algorithm.factorize import SVD
svd = SVD()
svd.set_data(train)
svd.compute(k=50, min_values=5, pre_normalize=None, mean_center=True, post_normalize=True,savefile='data/svdDataScores')
```

```python
# Evaluate test set using RMSE 
from recsys.evaluation.prediction import RMSE, MAE
rmse = RMSE()
mae = MAE()
for rating, user_id, item_id in test.get():
    try:
        pred_rating = svd.predict(user_id, item_id)
        rmse.add(rating, pred_rating)
    except KeyError:
        continue
        
score_max = 10.; score_min = 1.; 
print 'RMSE=%s' % rmse.compute()
print 'NRMSE=%s' % (rmse.compute()/(score_max-score_min))
print "Netflix's Prize Winning NRMSE=%s" % (0.8563/4.)
```

Here it is seen that this recommendation system has a normalized RMSE of 0.31 and is therefore performing very well. In order to put this into perspective, the million dollar prize winning Netflix's NRMSE was 0.21.
